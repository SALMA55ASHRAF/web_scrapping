```python
# web scrapping
#task 1 = collect all url linked to form that page
from bs4 import BeautifulSoup
import csv
import requests
import pandas as pd
url="https://www.house.gov/representatives"
page=requests.get(url)
src=page.content
soup=BeautifulSoup(src,"lxml")
my_table=soup.find('div',{'class':"view-content"})
tabl1=soup.find('table',{'id':'housegov_reps_by_state-block_default-514225866'})
#name_of1st_table=tabl1.caption.text

name_of_cols=list(soup.find('tr').text.strip().split("\n"))
df=pd.DataFrame(columns=name_of_cols)
number_of_rows=soup.find_all("td",{'headers':"view-value-2-table-column"})
number_of_rows_list=[]
for i in number_of_rows:
  number_of_rows_list.append(i.text.strip())
df['District']=number_of_rows_list
list_of_number_of_rows=[]
for i in number_of_rows:
  list_of_number_of_rows.append(i.text.strip())
names_in_name_col=soup.find_all("td",{'headers':'view-value-4-table-column'})
all_urls=[]
for td in names_in_name_col:
  for a in td('a'):
    all_urls.append(a.text.strip())
df['Name']=all_urls
#df['Links']=
party=soup.find_all('td',{'headers':'view-value-7-table-column'})
party_list=[]
for party in party:
  party_list.append(party.text.strip())
df['Party']=party_list
office_room=soup.find_all('td',{'headers':'view-value-9-table-column'})
office_room_list=[]
for i in office_room:
  office_room_list.append(i.text.strip())
df['Office Room']=office_room_list
phone_numbers=soup.find_all('td',{'headers':'view-value-10-table-column'})
phone_numbers_list=[]
for i in phone_numbers:
  phone_numbers_list.append(i.text.strip())
df['Phone']=phone_numbers_list
commite_ass=soup.find_all('td',{'headers':'view-markup-table-column'})
commite_ass_list=[]
for i in commite_ass:
  commite_ass_list.append(i.text.strip())
df['Committee Assignment']=commite_ass_list
with open('my_book_data_science_project.csv','w',encoding='utf-8') as f:
  dict_writer=csv.DictWriter(f,df.columns.tolist(),delimiter=',')
  dict_writer.writeheader()
  dict_writer.writerows(df.to_dict('records'))

links=[]
for d in soup('td',{'headers':'view-value-4-table-column'}):
  for a in d('a'):
    links.append(a.get("href"))
df['Links']=links
df
```
